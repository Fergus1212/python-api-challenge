```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
import scipy.stats as st
from scipy.stats import linregress
import os
from dotenv import load_dotenv
load_dotenv()

# Import API key

API_KEY = os.getenv('weather_api_key') 

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# Output File (CSV)
output_data_file = "output_data/cities.csv"

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

## Generate Cities List

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(low=-90.000, high=90.000, size=1500)
lngs = np.random.uniform(low=-180.000, high=180.000, size=1500)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

# Print the city count to confirm sufficient count
len(cities)
```

```python
city_name = []
clouds = []
country = []
wind_speed = []
date = []
latitude = []
longitude = []
max_temp = []
humidity = []

counter = 1 
print(f"Beginning Data Retrieval")
print(f"-------------------------------")
url = "http://api.openweathermap.org/data/2.5/weather?units=Imperial&APPID=" + API_KEY
for city in cities:  
    
    try: 
        response = requests.get(f"{url}&q={city}").json() 
        city_name.append(response["name"])
        clouds.append(response["clouds"]["all"])
        country.append(response["sys"]["country"])
        date.append(response["dt"])
        humidity.append(response["main"]["humidity"])
        max_temp.append(response["main"]["temp_max"])
        latitude.append(response["coord"]["lat"])
        longitude.append(response["coord"]["lon"])
        wind_speed.append(response["wind"]["speed"])
        city_record = response["name"]
        print(f"Processing Record {counter} | {city_record}")
        
        counter= counter + 1
        
        time.sleep(1.01)
        
    # If no record found "skip" to next call
    except:
        print("City not found. Skipping...")
    continue
```

### Convert Raw Data to DataFrame
* Export the city data into a .csv.
* Display the DataFrame

```python
city_data = pd.DataFrame({'City':city_name,
                         'Cloudiness':clouds,
                         'Country':country,
                         'Date':date,
                         'Humidity':humidity,
                         'Lat':latitude,
                         'Lng':longitude,
                         'Max Temp':max_temp,
                         'Wind Speed':wind_speed})
city_data
```

```python
city_data.to_csv('../output_data/cities.csv')
```

```python
# Create Northern and Southern Hemisphere DataFrames
north_city = city_data[city_data['Lat']>=0]
south_city = city_data[city_data['Lat']<0]
```

####  Northern Hemisphere - Max Temp vs. Latitude Linear Regression

```python
correlation = round(st.pearsonr(north_city['Max Temp'],north_city['Lat'])[0],2)
print(f"The correlation coefficient between Max Temp and Lat is {correlation}")
(slope, intercept, rvalue, pvalue, stderr) = linregress(north_city['Max Temp'], north_city['Lat'])
regress_values = north_city['Lat'] * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.scatter(north_city['Max Temp'], north_city['Lat'])
plt.plot(north_city['Lat'],regress_values,"r-")
plt.annotate(line_eq,(42,16),fontsize=15,color="red")
plt.title('Northern Hemisphere - Max Temp vs. Latitude Linear Regression')
plt.xlabel('Max Temp')
plt.ylabel('Lat')
```
